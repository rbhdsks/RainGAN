```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Data for the graphs
models = ['DCGAN', 'Pix2Pix', 'StyleGAN', 'SyRaGAN']
fid_scores = [18.45, 15.32, 10.57, 8.34]
lpips_scores = [0.29, 0.24, 0.18, 0.12]

# Set up the style
sns.set_style("whitegrid")
plt.rcParams.update({'font.size': 12})

# Create a figure with subplots for side-by-side bar charts
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# Bar width
bar_width = 0.4

# X locations for the groups
x = np.arange(len(models))

# Plotting FID scores
ax[0].bar(x - bar_width / 2, fid_scores, width=bar_width, color=sns.color_palette("Blues", len(models)))
ax[0].set_title('FID Scores for Different GAN Models')
ax[0].set_xlabel('GAN Model')
ax[0].set_ylabel('FID Score')
ax[0].set_xticks(x)
ax[0].set_xticklabels(models)
ax[0].set_ylim(0, max(fid_scores) + 5)
ax[0].grid(axis='y', linestyle='--', alpha=0.7)

# Adding value annotations
for i, v in enumerate(fid_scores):
    ax[0].text(i - bar_width / 2, v + 0.5, f'{v:.2f}', ha='center')

# Plotting LPIPS scores
ax[1].bar(x + bar_width / 2, lpips_scores, width=bar_width, color=sns.color_palette("Reds", len(models)))
ax[1].set_title('LPIPS Scores for Different GAN Models')
ax[1].set_xlabel('GAN Model')
ax[1].set_ylabel('LPIPS Score')
ax[1].set_xticks(x)
ax[1].set_xticklabels(models)
ax[1].set_ylim(0, max(lpips_scores) + 0.1)
ax[1].grid(axis='y', linestyle='--', alpha=0.7)

# Adding value annotations
for i, v in enumerate(lpips_scores):
    ax[1].text(i + bar_width / 2, v + 0.01, f'{v:.2f}', ha='center')

# Adjust layout
plt.tight_layout()
plt.show()
```
